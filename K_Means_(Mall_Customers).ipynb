```python
# import dataset

import pandas as pd
dataset = pd.read_csv('Datasets/mall_customers.csv')
```

```python
# spliting columns

x = dataset.iloc[:, [3, 4]].values
```

```python
# plotting wcss curve and marking elbow

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss = [] #empty array
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('No.of clusters')
plt.ylabel('WCSS \ Inertia')

#The bend (or elbow) here at K = 5 indicates the optimal number of cluster
plt.annotate('Elbow', xy=(5, 50000), xytext=(6, 100000), arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

```python
# prediction

kmeans = KMeans(n_clusters = 5, init = "k-means++", random_state=42)
y_pred = kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_pred == 0,0], x[y_pred == 0,1], s = 60, c = 'red', label = 'Cluster 1')  

#  Here, X[y_pred==0,0] is X axis and X[y_pred==0,1] is Y axis
#  We're plotting scatters for cluster=0 i.e, our first cluster , in red color

plt.scatter(x[y_pred == 1,0], x[y_pred == 1,1], s = 60, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_pred == 2,0], x[y_pred == 2,1], s = 60, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_pred == 3,0], x[y_pred == 3,1], s = 60, c = 'violet', label = 'Cluster 4')
plt.scatter(x[y_pred == 4,0], x[y_pred == 4,1], s = 60, c = 'yellow', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'black', label = 'Centroids')
plt.legend()
plt.show()
```
